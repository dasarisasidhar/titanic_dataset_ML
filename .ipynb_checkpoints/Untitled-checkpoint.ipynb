```python
import numpy as np
import pandas as pd
import os
import matplotlib
```

```python
path = os.path.abspath('..')
print(path)
```

```python

dataset_path = os.path.join(path, 'datasets',"titanic")
train_file_path = os.path.join(dataset_path, 'train.csv') 
test_file_path = os.path.join(dataset_path, 'test.csv')
print(dataset_path)
```

```python
train_data_frame = pd.read_csv(train_file_path, index_col='PassengerId')
test_data_frame = pd.read_csv(test_file_path, 'PassengerId')
```

```python
#type(train_data_frame)  #to check the type
#train_data_frame.info() #to get info
#test_data_frame.info()
#print(train_data_frame)#to see data how it looks
#print(test_data_frame)
```

```python
test_data_frame['Survived'] = -888  # adding Survived with default value to test df
```

```python
df = pd.concat((train_data_frame, test_data_frame), axis = 0) #axis is your to concat row wise
```

```python
# gives description only for numerical not catogarical
df.describe() #to get base staticial details mean median, ...
```

```python
# for numerical we can perform
print("min fare : {0}".format(df.Fare.min()))
print("max fare : {0}".format(df.Fare.max()))
print("fare range : {0}".format(df.Fare.max() - df.Fare.min()))
print("25 % : {0}".format(df.Fare.quantile(.25)))
print("70 % : {0}".format(df.Fare.quantile(.70)))
print("variance fare: {0}".format(df.Fare.var()))
print("sd fare : {0}".format(df.Fare.std()))
```
